# Taking important features in a single dataframe to get their keywords out and store in one feature called `tags`, which further will be used in segmentation of the recommender system.


```python
# Important features for categorizing movie types
# geners
# id
# keywords
# title
# overview
# cast
# crew
```

```python
movies = movies[['movie_id','title','genres','keywords','overview','cast','crew']]
```

```python
import ast
```

```python
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(4)
```

## Keywords

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(4)
```

## Taking top 3 from the cast ,they are are lead actors and actorresses of the movie.

```python
movies.cast
```

```python
def convert3(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(3)
```

## Taking out the DIRECTOR from the crew

```python
movies['crew'][0]
```

```python
def fetch_job(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      # counter+=1
      break
  return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_job)
```

```python
movies.head(3)
```

## Overview

```python
movies['overview'][0]
```

```python
movies.dropna(inplace=True)
```

```python
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies.head(3)
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(3)
```

# Merging together the tags into one feature `tags`

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(3)
```

## Creating `new_df` with features movie_idm, `title`, `tags`

```python
new_df = movies[['movie_id','title','tags']]
```

## Joining the `tags` into one single string of `tags` feature

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

# Converting the  the tags string

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head(5)
```

```python
new_df['tags'][0]
```

```python
new_df['tags'][1]
```

```python
import nltk
```

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(wnl.lemmatize(i,'v'))

  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
stem(new_df['tags'][0])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

## Converting the tags into countvectors

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

## Getting cosine_similarity of the vectors

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])), reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  # print(movie_index)
  distances = similarity[movie_index]
  # print(distances)
  movies_list = sorted(list(enumerate(distances)), reverse=True,key=lambda x:x[1])[1:6]
  # print(movies_list)

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
